# Problem statement
Collect raw Instagram data 

Answer these questions :-
- Who has max posts
- Who has max followers
- Who follows max peoples
- How many categories (Digital creators, Non profitable foundation etc. Do we have?)

```python
with open("data.txt") as f:
    data=f.read()
```

```python
chunks=data.split("\n\n")
chunks=[c for c in chunks if len(c)>5]
```

```python
def parse_chunk(chunk):
    chunk=chunk.strip()
    sep_chunk=chunk.split('\n')
    username=sep_chunk[0]
    no_of_post=int(sep_chunk[1].split(' post')[0].replace(",",""))
    followers=float((sep_chunk[2].split(' ')[0].replace(",","")).replace('K',"").replace('M',""))
    if "K" in sep_chunk[2]:
        followers=int(followers*1000)
    elif "M" in sep_chunk[2]:
        followers=int(followers*1000000)
    else:
        followers=int(followers)
    followings=float(sep_chunk[3].split(' ')[0].replace(",","").replace('K',"").replace('M',""))
    if "K" in sep_chunk[3]:
        followings=int(followers*1000)
    elif "M" in sep_chunk[3]:
        followings=int(followers*1000000)
    else:
        followings=int(followings)
    name=sep_chunk[4]
    if len(sep_chunk)>5:
        type_of_page=sep_chunk[5]
        bio="\n".join(sep_chunk[6:])
    else:
        type_of_page="Unknown"
        bio=""
        
    # print(username, no_of_post, followers,followings, name,type_of_page, bio, sep='\n')
    return {'username': username, "no_of_post": no_of_post, "followers": followers,'followings': followings, 'name': name,"type_of_page": type_of_page, "bio": bio}
# print(parse_chunk(chunks[0]))
```

```python
all_chunks=[parse_chunk(chunk) for chunk in chunks]
    
```

```python
import json
```

```python
s=json.dumps(all_chunks,indent=4)
with open ("data.json",'w') as fjson:
    fjson.write(s)
```

## Who has max posts?

```python
max_post=0
for chunk in all_chunks:
    if max_post<=chunk["no_of_post"]:
        max_post=chunk["no_of_post"]
        chunks_with_max_post=chunk
print(chunks_with_max_post)
```

## Who has max followers?

```python
max_followers=0
for chunk in all_chunks:
    if max_followers<=chunk["followers"]:
        max_followers=chunk["followers"]
        chunks_with_max_followers=chunk
print(chunks_with_max_followers)
```

## Who follows max peoples

```python
max_followings=0
for chunk in all_chunks:
    if max_followings<=chunk["followings"]:
        max_followings=chunk["followings"]
        chunks_with_max_followings=chunk
print(chunks_with_max_followings)
```

## How many categories (Digital creators, Non profitable foundation etc. Do we have?)

```python
categories=set()
for chunk in all_chunks:
    categories.add(chunk['type_of_page'])
print(f"There are {len(categories)} categories")
```
