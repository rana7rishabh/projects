```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        return json.load((f))
```

```python
def find_people_you_may_know(user_id, data):
    user_friends={}
    for user in data['users']:
        user_friends[user['id']]=set(user['friends']) #unique friends in the dictionary user_friends

    if user_id not in user_friends:  #user in not in friend list then return empty list
        return[]
    
    direct_friends=user_friends[user_id]
    suggestions={}
    for friend in direct_friends:
        for mutual in user_friends[friend]:
            if mutual!=user_id and mutual not in direct_friends: #wo khud na ho aur uske friend list pehle se wo dost na ho
                #count suggestions
                suggestions[mutual]=suggestions.get(mutual,0)+1
    sorted_suggestions=sorted(suggestions.items(),key= lambda x:x[1], reverse=True)
    return [user_id for user_id, mutual_count in sorted_suggestions]
```

```python
#Load the data
data=load_data('massive_data.json')
```

```python
user_id=10
```

```python
recc=find_people_you_may_know(user_id, data)
```
